```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

data_before_cleaning = pd.read_csv("Uncleaned Data.csv", index_col = "date")
data_after_cleaning = pd.read_csv("NewCombined.csv", index_col= "date")
```

```python
# 1. Distribution of Variables
data_before_cleaning[['tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres']].hist(bins=20, figsize=(10, 8), layout=(2, 3))
plt.suptitle('Distribution of Weather Variables')
plt.show()
```

```python
# 1. Distribution of Variables
data_after_cleaning[['tavg', 'tmin', 'tmax', 'prcp', 'wspd', 'pres']].hist(bins=20, figsize=(10, 8), layout=(2, 3))
plt.suptitle('Distribution of Weather Variables')
plt.show()
```

```python
# 4. Scatter Plot between tmin, tmax, and other variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tmin', y='tmax', data=data_before_cleaning, hue='wspd', palette='coolwarm')
plt.title('Scatter Plot of Tmin vs Tmax (Colored by Windspeed)')
plt.xlabel('Min Temperature (C)')
plt.ylabel('Max Temperature (C)')
plt.show()
```

```python
# 4. Scatter Plot between tmin, tmax, and other variables
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tmin', y='tmax', data=data_after_cleaning, hue='wspd', palette='coolwarm')
plt.title('Scatter Plot of Tmin vs Tmax (Colored by Windspeed)')
plt.xlabel('Min Temperature (C)')
plt.ylabel('Max Temperature (C)')
plt.show()
```
